```python
import pandas as pd
import seaborn as sns
```

```python
dados = pd.read_csv('tips_atualizado.csv')
dados.head()
```

```python
dados.describe()
```

```python
dados.columns = ['valor_da_conta', 'gorjeta', 'sobremesa', 'dia_da_semana', 'hora_do_dia', 'total_de_pessoas']
dados.head()
```

```python
gorjetas = dados
gorjetas.head()
```

```python
modification = {
    'No': 'Não',
    'Yes': 'Sim'
}
gorjetas['sobremesa'] = gorjetas['sobremesa'].map(modification)
gorjetas.head()
```

```python
gorjetas['dia_da_semana'].unique()
```

```python
modification = {
    'Sun': 'Domingo',
    'Sat': 'Sábado',
    'Thur': 'Quinta',
    'Fri': 'Sexta'
}
gorjetas['dia_da_semana'] = gorjetas['dia_da_semana'].map(modification)
gorjetas.head()
```

```python
gorjetas['hora_do_dia'].unique()
```

```python
modification = {
    'Dinner': 'Jantar',
    'Lunch': 'Almoço'
}
gorjetas['hora_do_dia'] = gorjetas['hora_do_dia'].map(modification)
gorjetas.head()
```

```python
sns.heatmap(gorjetas.corr(), annot=True)
```

```python
sns.scatterplot(x = 'valor_da_conta', y = 'gorjeta', data = gorjetas)
```

```python
gorjetas.info()
```

```python
gorjetas['porcentagem'] =  gorjetas['gorjeta'] / gorjetas['valor_da_conta']
gorjetas['porcentagem'] =  gorjetas['porcentagem'].round(2)
gorjetas.head()
```

```python
sns.scatterplot(x = 'valor_da_conta', y = 'porcentagem', data = gorjetas)
```

```python
sns.relplot(x = 'valor_da_conta', y = 'porcentagem', kind = 'line', data = gorjetas)
```

```python
sns.lmplot(x = 'valor_da_conta', y = 'porcentagem', data = gorjetas)
```

```python
gorjetas[gorjetas['sobremesa'] == 'Sim'].describe()
```

```python
sns.catplot(x = 'sobremesa', y = 'gorjeta', data = gorjetas)
```

```python
sns.relplot(x = 'valor_da_conta', y = 'gorjeta', hue = 'sobremesa', data = gorjetas)
```

```python
sns.relplot(x = 'valor_da_conta', y = 'gorjeta', hue = 'sobremesa', col = 'sobremesa', data = gorjetas)
```

```python
sns.lmplot(x = 'valor_da_conta', y = 'gorjeta', col = 'sobremesa', hue = 'sobremesa', data = gorjetas)
```

```python
sns.lmplot(x = 'valor_da_conta', y = 'porcentagem', col = 'sobremesa', hue = 'sobremesa', data = gorjetas)
```

```python
sns.relplot(x = 'valor_da_conta', y = 'gorjeta', hue = 'sobremesa', col = 'sobremesa', kind = 'line', data = gorjetas)
```

# Teste de Hipótese

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'")['porcentagem']
sem_sobremesa = gorjetas.query("sobremesa == 'Não'")['porcentagem']
```

```python
ranksums(sobremesa, sem_sobremesa)
```
